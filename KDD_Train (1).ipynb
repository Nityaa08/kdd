```python
d["labels"] = d["labels"].replace(['neptune', 'warezclient', 'ipsweep', 'portsweep',
       'teardrop', 'nmap', 'satan', 'smurf', 'pod', 'back',
       'guess_passwd', 'ftp_write', 'multihop', 'rootkit',
       'buffer_overflow', 'imap', 'warezmaster', 'phf', 'land',
       'loadmodule', 'spy', 'perl'], 'attack')
```

```python
d["labels"].unique()
```

```python
x=d.iloc[:,:-1].values
y=d.iloc[:,41].values
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_x_1=LabelEncoder()
labelencoder_x_2=LabelEncoder()
labelencoder_x_3=LabelEncoder()
x[:,1]=labelencoder_x_1.fit_transform(x[:,1])
x[:,2]=labelencoder_x_2.fit_transform(x[:,2])
x[:,3]=labelencoder_x_3.fit_transform(x[:,3])
```

```python
display(set(x[:,1]))
display(set(x[:,2]))
display(set(x[:,3]))
```

```python
display(set(y))
```

```python
from sklearn.preprocessing import StandardScaler
x=StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=6)
```

```python
pct=pca.fit_transform(x)
```

```python
print("Eigenvalues")
print(pca.explained_variance_)
print("eigen vectors")
print(pca.components_)
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
plt.plot(pca.explained_variance_,marker='o')
plt.xlabel("eigen number")
plt.ylabel("eigensize")
plt.title("Scree plot")
```

```python
principal_x=pd.DataFrame(pct,columns=['pc1','pc2','pc3','pc4','pc5','pc6'])
```

```python
y=pd.DataFrame(y)
display(y)
```

```python
final=pd.concat([principal_x,y],axis=1)
```

```python
display(final)
```

```python
final.columns
```

```python
final.rename(columns={0:'target'},inplace=True)
```

```python
final.columns
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1',fontsize=15)
ax.set_ylabel('Principal Component 2',fontsize=15)
ax.set_title('2 component PCA',fontsize=20)
targets=['attack','normal']
colors=['r','g']
for target,color in zip(targets,colors):
    indicesToKeep=final['target']==target
    ax.scatter(final.loc[indicesToKeep,'pc1']
              ,final.loc[indicesToKeep,'pc2']
              ,c=color
              ,s=50)
ax.legend(targets)
ax.grid()
```
